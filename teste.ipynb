```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2+2*x+1


# criando intervalo de valores para x
x = np.linspace(-10, 10, 400)
y = f(x)

# plotando gáfico
plt.plot(x, y, label = "f(x) = x^2+2x+1")
plt.axhline(0, color='black', linewidth=0.8)
plt.axhline(0, color='black', linewidth=0.8)
plt.legend()
plt.grid(True)
plt.title("Gráfico da função real f(x)=x^2x+1")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
